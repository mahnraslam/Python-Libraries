```python
import pandas as pd
# DATA CLEANING

'''
By default, the dropna() method returns a new DataFrame, and will not change the original. inplace : True for permanently storing the changes inoriginal dataset
'''
#df = pd.read_csv('my_arr.txt', delim_whitespace=True, index_col=0)
df = pd.read_csv('data.csv', index_col = 0)
```

```python
# Convert 'Pulse' column to float64 and assign it back to the DataFrame
# df['Pulse'] = df['Pulse'].astype('float64')
df.Pulse.astype("float64", copy=True, errors='raise')
df.astype({'Pulse': float})
df.dtypes
```

```python
#For the selection of missing data 
df[pd.isnull(df.Calories)]
df.Date.isnull()
```

```python
# fillna()  for filing empty spaces
#.replace can be used to replace the null values
df.Calories.fillna('Uknown')
df.Calories.replace("Uknown","")
df
```

```python
#To drop enpty rows
#One way to deal with empty cells is to remove rows that contain empty cells. new df
new  =  df.DataFrame({"headers": ["student_id", "name", "age"], "values": [[470, "Quincy", 20], [300, "Liam", 15]]})
new_df = df.dropna()
new.sort_values(ascending=False)
```

```python
#To fill the empty with mean value
df.Calories = df.Calories.fillna(df.Calories.mean() )
df
```

```python
df["Date"] = pd.to_datetime(df['Date']) 
df.dropna(subset=['Date'], inplace=True)

res =df[df['Duration']>50].sort_values(by ='Duration', ascending = False).Calories
 
res.reset_index(drop=True)
```
